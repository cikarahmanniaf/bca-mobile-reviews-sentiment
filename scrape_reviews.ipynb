# Data Collection: Scraping Ulasan BCA Mobile di Google Play


Langkah pertama dalam proyek ini adalah mengumpulkan data ulasan pengguna aplikasi BCA Mobile dari platform Google Play Store. Data ulasan ini akan menjadi dasar untuk proses analisis sentimen. Proses pengambilan data dilakukan menggunakan teknik web scraping. Fokus scraping diarahkan pada ulasan dalam bahasa Indonesia. Dataset hasil scraping akan disimpan dalam format .csv, yang selanjutnya akan digunakan untuk tahap pelatihan model.

```python
# Mengimpor pustaka google_play_scraper
from google_play_scraper import app, reviews_all, Sort
```

```python
from google.colab import files # untuk mengunduh file ke lokal
```

```python
# Mengambil semua ulasan dari aplikasi dengan ID 'com.bca' di Google Play Store.

scrapreview = reviews_all(
    'com.bca',                # ID aplikasi BCA Mobile
    lang='id',                # Bahasa ulasan (default: 'en')
    country='id',             # Negara (default: 'us')
    sort=Sort.NEWEST,         # Urutan ulasan (default: Sort.MOST_RELEVANT)
    count=12000               # Jumlah maksimum ulasan yang ingin diambil
)
```

```python
# Menyimpan ulasan dalam file CSV
import csv

with open('bcamobile_review.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Menulis header kolom
    for review in scrapreview:
        writer.writerow([review['content']])  # Menulis konten ulasan ke dalam file CSV
```
